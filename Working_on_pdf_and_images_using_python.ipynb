```python
# Text extration from pdf and images
import fitz
import pytesseract
from PIL import Image
import os
```

```python
def extract_text_from_pdf(pdf_path):
    """Extracts text from a PDF file."""
    with fitz.open(pdf_path) as pdf:
        text = ""
        for page_num in range(len(pdf)):
            page = pdf[page_num]
            text += page.get_text("text")
    return text

def ocr_pdf(pdf_path):
    """OCR for image-based PDFs"""
    text = ""
    with fitz.open(pdf_path) as pdf:
        for page_num in range(len(pdf)):
            page = pdf.load_page(page_num)
            pix = page.get_pixmap()
            img_path = f"/tmp/page_{page_num}.png"
            pix.save(img_path)
            text += pytesseract.image_to_string(Image.open(img_path))
            os.remove(img_path)
    return text
```

```python
import spacy

# Load the SpaCy language model
nlp = spacy.load("en_core_web_sm")

def preprocess_text(text):
    """Clean and preprocess the text."""
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
    return " ".join(tokens)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from transformers import pipeline

# Initialize the transformer model for summarization
summarizer = pipeline("summarization")

def extract_keywords(text, top_n=10):
    """Extract top N keywords using TF-IDF."""
    vectorizer = TfidfVectorizer(stop_words='english', max_features=top_n)
    vectors = vectorizer.fit_transform([text])
    keywords = vectorizer.get_feature_names_out()
    return keywords

def summarize_text(text):
    """Summarize long text."""
    return summarizer(text, max_length=130, min_length=30, do_sample=False)[0]['summary_text']
```

```python
import sqlite3

def init_db():
    """Initialize the SQLite database."""
    conn = sqlite3.connect("pdf_reader.db")
    c = conn.cursor()
    c.execute('''CREATE TABLE IF NOT EXISTS pdfs (
                 id INTEGER PRIMARY KEY AUTOINCREMENT,
                 filename TEXT,
                 text TEXT,
                 summary TEXT,
                 keywords TEXT)''')
    conn.commit()
    conn.close()

def store_pdf_data(filename, text, summary, keywords):
    """Store the PDF data into SQLite."""
    conn = sqlite3.connect("pdf_reader.db")
    c = conn.cursor()
    c.execute("INSERT INTO pdfs (filename, text, summary, keywords) VALUES (?, ?, ?, ?)",
              (filename, text, summary, ",".join(keywords), summary))
    conn.commit()
    conn.close()
```

```python
def query_pdf_data(filename):
    """Query the database for a specific PDF."""
    conn = sqlite3.connect("pdf_reader.db")
    c = conn.cursor()
    c.execute("SELECT summary, keywords FROM pdfs WHERE filename = ?", (filename,))
    result = c.fetchone()
    conn.close()
    return result
```

```python
import streamlit as st

# Initialize the database
init_db()

# Streamlit UI
st.title("PDF Reader ML Pipeline")
uploaded_file = st.file_uploader("Upload your PDF file", type=["pdf"])

if uploaded_file is not None:
    # Extract text from PDF
    pdf_path = f"./{uploaded_file.name}"
    with open(pdf_path, "wb") as f:
        f.write(uploaded_file.read())

    # Extract text
    text = extract_text_from_pdf(pdf_path) or ocr_pdf(pdf_path)

    # Preprocess text
    clean_text = preprocess_text(text)

    # Perform keyword extraction and summarization
    keywords = extract_keywords(clean_text)
    summary = summarize_text(clean_text)

    # Store results in DB
    store_pdf_data(uploaded_file.name, clean_text, summary, keywords)

    # Display results
    st.subheader("Summary")
    st.write(summary)

    st.subheader("Keywords")
    st.write(keywords)

    # Query for the same file if needed
    if st.button('Query Database for This File'):
        result = query_pdf_data(uploaded_file.name)
        if result:
            st.write(f"Summary: {result[0]}")
            st.write(f"Keywords: {result[1]}")
        else:
            st.write("No data found.")
```
